```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

#   
matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False

# 
worker = pd.read_csv('foreign_worker.csv',encoding='euc-kr').iloc[3:]
resident = pd.read_csv('foreign_resident.csv',encoding='euc-kr').iloc[4:]

#worker plotting
x_worker=worker['()(1)']
y_worker=[]

for row in worker['2021']:
        y_worker.append((int)(row))
        
plt.figure(1)
plt.bar(x_worker, y_worker)
plt.xticks(rotation=70)


#resident plotting
x_resident=resident['(1)']
y_resident=[]

for row in resident['2021.11']:
    y_resident.append((int)(row))


plt.figure(2)
plt.bar(x_resident,y_resident)
plt.xticks(rotation=70)




plt.figure(3)
x=[1]
y=[1]

plt.bar(x,y)
```
