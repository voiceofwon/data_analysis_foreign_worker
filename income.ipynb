```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib


#그래프 한글 인코딩 설정
matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False

foreign = pd.read_csv('foreign_income.csv',encoding='euc-kr')

foreign_tax = foreign.iloc[58:59]

foreign_tax_personnel =[]

foreign_tax_personnel.append((int)(foreign_tax['2008']))
foreign_tax_personnel.append((int)(foreign_tax['2009']))
foreign_tax_personnel.append((int)(foreign_tax['2010']))
foreign_tax_personnel.append((int)(foreign_tax['2011']))
foreign_tax_personnel.append((int)(foreign_tax['2012']))
foreign_tax_personnel.append((int)(foreign_tax['2013']))
foreign_tax_personnel.append((int)(foreign_tax['2014']))
foreign_tax_personnel.append((int)(foreign_tax['2015']))
foreign_tax_personnel.append((int)(foreign_tax['2016']))
foreign_tax_personnel.append((int)(foreign_tax['2017']))
foreign_tax_personnel.append((int)(foreign_tax['2018']))
foreign_tax_personnel.append((int)(foreign_tax['2019']))
foreign_tax_personnel.append((int)(foreign_tax['2020']))
foreign_tax_personnel.append((int)(foreign_tax['2021']))


foreign_tax_amount =[]

foreign_tax_amount.append((int)(foreign_tax['2008.1']))
foreign_tax_amount.append((int)(foreign_tax['2009.1']))
foreign_tax_amount.append((int)(foreign_tax['2010.1']))
foreign_tax_amount.append((int)(foreign_tax['2011.1']))
foreign_tax_amount.append((int)(foreign_tax['2012.1']))
foreign_tax_amount.append((int)(foreign_tax['2013.1']))
foreign_tax_amount.append((int)(foreign_tax['2014.1']))
foreign_tax_amount.append((int)(foreign_tax['2015.1']))
foreign_tax_amount.append((int)(foreign_tax['2016.1']))
foreign_tax_amount.append((int)(foreign_tax['2017.1']))
foreign_tax_amount.append((int)(foreign_tax['2018.1']))
foreign_tax_amount.append((int)(foreign_tax['2019.1']))
foreign_tax_amount.append((int)(foreign_tax['2020.1']))
foreign_tax_amount.append((int)(foreign_tax['2021.1']))

foreign_tax_ratio = []

for val in range(14):
    foreign_tax_ratio.append(1000000 * foreign_tax_amount[val]/foreign_tax_personnel[val])
    

korean = pd.read_csv('korean_income.csv',encoding='euc-kr')

korean_tax = korean.iloc[58:59]

korean_tax_personnel =[]

korean_tax_personnel.append((int)(korean_tax['2008']))
korean_tax_personnel.append((int)(korean_tax['2009']))
korean_tax_personnel.append((int)(korean_tax['2010']))
korean_tax_personnel.append((int)(korean_tax['2011']))
korean_tax_personnel.append((int)(korean_tax['2012']))
korean_tax_personnel.append((int)(korean_tax['2013']))
korean_tax_personnel.append((int)(korean_tax['2014']))
korean_tax_personnel.append((int)(korean_tax['2015']))
korean_tax_personnel.append((int)(korean_tax['2016']))
korean_tax_personnel.append((int)(korean_tax['2017']))
korean_tax_personnel.append((int)(korean_tax['2018']))
korean_tax_personnel.append((int)(korean_tax['2019']))
korean_tax_personnel.append((int)(korean_tax['2020']))
korean_tax_personnel.append((int)(korean_tax['2021']))

korean_tax_amount = []

korean_tax_amount.append((int)(korean_tax['2008.1']))
korean_tax_amount.append((int)(korean_tax['2009.1']))
korean_tax_amount.append((int)(korean_tax['2010.1']))
korean_tax_amount.append((int)(korean_tax['2011.1']))
korean_tax_amount.append((int)(korean_tax['2012.1']))
korean_tax_amount.append((int)(korean_tax['2013.1']))
korean_tax_amount.append((int)(korean_tax['2014.1']))
korean_tax_amount.append((int)(korean_tax['2015.1']))
korean_tax_amount.append((int)(korean_tax['2016.1']))
korean_tax_amount.append((int)(korean_tax['2017.1']))
korean_tax_amount.append((int)(korean_tax['2018.1']))
korean_tax_amount.append((int)(korean_tax['2019.1']))
korean_tax_amount.append((int)(korean_tax['2020.1']))
korean_tax_amount.append((int)(korean_tax['2021.1']))


korean_tax_ratio =[]

for val in range(14):
   korean_tax_ratio.append(1000000*korean_tax_amount[val]/korean_tax_personnel[val]) 



year = []

for val in range(2008,2022):
    year.append(val)
    
    

```

```python
import statsmodels.api as sm

korean_tax_ratio_m = sm.add_constant(korean_tax_ratio)

model_0 = sm.OLS(foreign_tax_ratio,korean_tax_ratio_m).fit()
pred_0 = model_0.predict(korean_tax_ratio_m)

print(model_0.summary())

plt.figure(1)

plt.title('근로소득 세액공제 외국인/한국인')
plt.plot(korean_tax_ratio,foreign_tax_ratio,'o')
plt.plot(korean_tax_ratio,pred_0)
```

```python
plt.figure(1)
plt.title('한국인 외국인 인당 공제액')
plt.bar(year,korean_tax_ratio)
plt.bar(year,foreign_tax_ratio)
```
